```python
import pandas as pd
import numpy as np
import labellines
import matplotlib.pyplot as plt
import seaborn as sns
NUM_COLORS = 6
cm = plt.get_cmap('Paired')
colors = [cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)]
```

## TMT

```python
# smal version
df =pd.read_csv("../shared_data/percentiles_chart_boys.csv", header=0)

plt.figure(figsize = (12.5,5))
ax = plt.subplot()
idx = 1
category_data = df
color = '#016FA0'
ax.fill_between(category_data['x'], y1=category_data['X25'],
                    y2=category_data['X75'],
                    alpha=0.75, facecolor=color)

ax.fill_between(category_data['x'], 
                    y1=category_data['X10'],
                    y2=category_data['X90'],
                    alpha=0.4, facecolor=color)

ax.fill_between(category_data['x'], 
                    y1=category_data['X3'],
                    y2=category_data['X97'],
                    alpha=0.25, facecolor=color)


ax.set_ylabel('iTMT [MM]')
ax.set_xticks(range(4,31,1))
ax.set_yticks(range(6,21,1))
ax.set_xlim([4, 31])
ax.set_ylim([6, 21])
ax.set_xlabel('AGE [YEARS]')
ax.set_title('TEMPORALIS MUSCLE THICKNESS GROWTH CHART (MALE)')

p1 = ax.plot(category_data['x'], category_data['X50'], color='k')
p2 = ax.fill(np.NaN, np.NaN, facecolor=color,  alpha=0.75)
p3 = ax.fill(np.NaN, np.NaN, facecolor=color,  alpha=0.4)
p4 = ax.fill(np.NaN, np.NaN, facecolor=color,  alpha=0.25)

ax.grid(color = 'gray',which='major', linestyle = '-', linewidth = 0.5)
ax.grid(color = 'gray',which='minor', linestyle = '--', linewidth = 0.25)
ax.minorticks_on()

ax.plot(category_data['x'], category_data['X3'], linewidth=1, label="3",color=color,alpha=0.0)
ax.plot(category_data['x'], category_data['X10'], linewidth=1, label="10",color=color,alpha=0.0)
ax.plot(category_data['x'], category_data['X25'], linewidth=1, label="25",color=color,alpha=0.0)
ax.plot(category_data['x'], category_data['X50'], linewidth=1, label="50",color='k',alpha=1.0)
ax.plot(category_data['x'], category_data['X75'], linewidth=1, label="75",color=color,alpha=0.0)
ax.plot(category_data['x'], category_data['X90'], linewidth=1, label="90",color=color,alpha=0.0)
ax.plot(category_data['x'], category_data['X97'], linewidth=1, label="97",color=color,alpha=0.0)
labellines.labelLines(ax.get_lines(), align=False, color="k",xvals=[20,21,22,23,24,25,26])

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(15)
ax.minorticks_on()
```
