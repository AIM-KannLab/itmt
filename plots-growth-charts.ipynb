```python
import pandas as pd
import numpy as np
import labellines
import matplotlib.pyplot as plt
import seaborn as sns
NUM_COLORS = 6
cm = plt.get_cmap('Paired')
colors = [cm(1.*i/NUM_COLORS) for i in range(NUM_COLORS)]
```

## TMT

```python
df =pd.read_csv("data/percentiles_chart_boys.csv", header=0)

'''SMALL_SIZE = 18
MEDIUM_SIZE = 20
BIGGER_SIZE = 24

plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE) '''

plt.figure(figsize = (16,6))
ax = plt.subplot()
idx = 1
category_data = df

ax.fill_between(category_data['x'], y1=category_data['X25'],
                    y2=category_data['X75'],
                    alpha=0.75, facecolor=colors[idx])

ax.fill_between(category_data['x'], 
                    y1=category_data['X10'],
                    y2=category_data['X90'],
                    alpha=0.4, facecolor=colors[idx])

ax.fill_between(category_data['x'], 
                    y1=category_data['X3'],
                    y2=category_data['X97'],
                    alpha=0.25, facecolor=colors[idx])


ax.set_ylabel('TEMPORALIS MUSCLE THICKNESS [MM]')
ax.set_xticks(range(4,36,1))
ax.set_yticks(range(6,21,1))
ax.set_xlim([4, 35])
ax.set_ylim([6, 21])
ax.set_xlabel('AGE [YEARS]')
ax.set_title('TEMPORALIS MUSCLE THICKNESS GROWTH CHART (MALE)')

p1 = ax.plot(category_data['x'], category_data['X50'], color='k')
p2 = ax.fill(np.NaN, np.NaN, facecolor=colors[1],  alpha=0.75)
p3 = ax.fill(np.NaN, np.NaN, facecolor=colors[1],  alpha=0.4)
p4 = ax.fill(np.NaN, np.NaN, facecolor=colors[1],  alpha=0.25)
ax.legend([(p1[0]),(p2[0]),(p3[0]),(p4[0])], ["Median",'25%-75%',"10%-90%","3%-97%"])
ax.tick_params(labelbottom=True, labeltop=False, labelleft=True, labelright=True,
                     bottom=True, top=False, left=True, right=True)

ax.grid(color = 'gray',which='major', linestyle = '-', linewidth = 0.5)
ax.grid(color = 'gray',which='minor', linestyle = '--', linewidth = 0.25)
ax.minorticks_on()

ax.plot(category_data['x'], category_data['X3'], linewidth=1, label="3",color=colors[1],alpha=0.0)
ax.plot(category_data['x'], category_data['X10'], linewidth=1, label="10",color=colors[1],alpha=0.0)
ax.plot(category_data['x'], category_data['X25'], linewidth=1, label="25",color=colors[1],alpha=0.0)
ax.plot(category_data['x'], category_data['X50'], linewidth=1, label="50",color='k',alpha=1.0)
ax.plot(category_data['x'], category_data['X75'], linewidth=1, label="75",color=colors[1],alpha=0.0)
ax.plot(category_data['x'], category_data['X90'], linewidth=1, label="90",color=colors[1],alpha=0.0)
ax.plot(category_data['x'], category_data['X97'], linewidth=1, label="97",color=colors[1],alpha=0.0)
labellines.labelLines(ax.get_lines(), align=False, color="k",xvals=[30,30,30,30,30,30,30])

for item in ([ax.title, ax.xaxis.label, ax.yaxis.label] +
             ax.get_xticklabels() + ax.get_yticklabels()):
    item.set_fontsize(20)
```
