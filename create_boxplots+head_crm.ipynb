```python
bmis = []
df_ant = pd.read_csv("data/ABCD-studies/abcd_bmi_w_estimate.csv", header=0)
for i in range(0,len(df[(df['TMT PRED AVG']>15)])):
    if "NDAR" in df[(df['TMT PRED AVG']>15)]['ID'].iloc[i]:
        ids=df[(df['TMT PRED AVG']>15)]['ID'].iloc[i].split("sub-")[1].split("_")[0]
        ses_id=df[(df['TMT PRED AVG']>15)]['ID'].iloc[i].split("ses-")[1].split("_")[0].lower()
        bmi = df_ant[(df_ant['id']==ids) & (df_ant['Visit']==ses_id)]['BMI']
        if len(bmi)>0:
            bmis.append(bmi)
print("Average BMI for the TMT>15mm is", round(np.average(bmis),2),"n=",len(bmis))
```

## Head crm

```python
input_annotation_file = 'data/dataset_measured_heads_filtered.csv'
df = pd.read_csv(input_annotation_file, header=0,)
```

```python
df
```

```python
df['Adjusted TMT by head crm'] = df['TMT PRED AVG filtered']/df['perimeter_opencv']
df=df[df["Age"]<35]
```

```python
#plot general
sns.set(rc={'figure.figsize':(16.7,6.27),'axes.facecolor':'white', 'figure.facecolor':'white'})
sns.set_style("whitegrid")
df['Sex'] = df['Gender'].map({2:'Female',1:'Male'})
sns.boxplot(x="Age", y="TMT PRED AVG filtered", hue="Sex", data=df, showfliers=False,
            palette="colorblind", width=0.5).set_title("Biological gender wrt TMT n="+str(df.shape[0]))
#plt.savefig('data/abcd_plots/gender.png', dpi=300)
```

```python
#plot general
sns.set(rc={'figure.figsize':(16.7,6.27),'axes.facecolor':'white', 'figure.facecolor':'white'})
sns.set_style("whitegrid")
df['Sex'] = df['Gender'].map({2:'Female',1:'Male'})
sns.boxplot(x="Age", y="Adjusted TMT by head crm", hue="Sex", data=df, showfliers=False,
            palette="colorblind", width=0.5).set_title("Adjusted TMT by head crm n="+str(df.shape[0]))
plt.savefig('data/abcd_plots/crm.png', dpi=300)
```

```python
from scipy.stats import pearsonr
corr, _ = pearsonr(df['TMT PRED AVG filtered'], df['perimeter_opencv'])
print('Pearsons correlation: %.3f' % corr)

from scipy.stats import spearmanr
corr, _ = spearmanr(df['TMT PRED AVG filtered'], df['perimeter_opencv'])
print('Spearmans correlation: %.3f' % corr)
```

```python
#plot general
sns.set(rc={'figure.figsize':(16.7,6.27),'axes.facecolor':'white', 'figure.facecolor':'white'})
sns.set_style("whitegrid")
df['Sex'] = df['Gender'].map({2:'Female',1:'Male'})
sns.boxplot(x="Age", y="perimeter_opencv", hue="Sex", data=df, showfliers=False,
            palette="colorblind", width=0.5).set_title("Head perimeter (opencv) n="+str(df.shape[0]))
plt.savefig('data/abcd_plots/perimeter_opencv.png', dpi=300)
```
